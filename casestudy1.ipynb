````python
import sqlite3
df = pd.read_csv('/content/cleaned_data.csv')
#connect to a database
conn = sqlite3.connect("Any_Database_Name.db") #if the db does not exist, this creates a Any_Database_Name.db file in the current directory
#store your table in the database:

# Drop the 'Index' column if it exists
if 'Index' in df.columns:
    df = df.drop('Index', axis=1)

df.to_sql('accidents', conn, if_exists='replace') # Use if_exists='replace' to overwrite the table if it exists
````

````python
sql_string = 'SELECT * FROM accidents'
df = pd.read_sql(sql_string, conn)
````

1. Most Common Types of Vehicles Involved in Accidents


````python
query1 = """
SELECT Vehicle_Type, COUNT(*) AS count
FROM accidents
GROUP BY Vehicle_Type
ORDER BY count DESC;
"""
vehicle_type_counts = pd.read_sql_query(query1, conn)
vehicle_type_counts
````

2. Most Frequent Weather Conditions Associated with Accidents


````python
query2 = """
SELECT Weather_Conditions, COUNT(*) AS count
FROM accidents
GROUP BY Weather_Conditions
ORDER BY count DESC;
"""
weather_condition_counts = pd.read_sql_query(query2, conn)
weather_condition_counts
````

3. Number of Casualties Vary Across Different Types of Accidents


````python
query3 = """
SELECT Accident_Severity,
       AVG(Number_of_Casualties) AS avg_casualties,
       MIN(Number_of_Casualties) AS min_casualties,
       MAX(Number_of_Casualties) AS max_casualties
FROM accidents
GROUP BY Accident_Severity;
"""
casualties_by_severity = pd.read_sql_query(query3, conn)
casualties_by_severity
````

4. Geographic Areas with the Highest Occurrence of Accidents


````python
query4 = """
SELECT [District Area], COUNT(*) AS count
FROM accidents
GROUP BY [District Area]
ORDER BY count DESC;
"""
district_area_counts = pd.read_sql_query(query4, conn)
district_area_counts
````

5. Impact of Time of Day on the Frequency and Severity of Accidents
*italicised text*[link text](https://)

````python
query5 = """
SELECT strftime('%H', Accident_Date) AS hour,
       COUNT(*) AS count,
       AVG(Number_of_Casualties) AS avg_casualties
FROM accidents
GROUP BY hour
ORDER BY hour;
"""
time_of_day_impact = pd.read_sql_query(query5, conn)
time_of_day_impact
````

6. Distribution of Accidents Across Different Days of the Week


````python
query6 = """
SELECT strftime('%w', Accident_Date) AS day_of_week,
       COUNT(*) AS count
FROM accidents
GROUP BY day_of_week
ORDER BY day_of_week;
"""
day_of_week_distribution = pd.read_sql_query(query6, conn)
day_of_week_distribution
````

7. Light Conditions Associated with Severe Accidents


````python
query7 = """
SELECT Light_Conditions,
       COUNT(*) AS count,
       SUM(CASE WHEN Accident_Severity = 'Serious' THEN 1 ELSE 0 END) AS serious_accidents
FROM accidents
GROUP BY Light_Conditions
ORDER BY serious_accidents DESC;
"""
light_conditions_severity = pd.read_sql_query(query7, conn)
light_conditions_severity
````

8. Correlation Between the Number of Vehicles Involved and the Severity of Accidents


````python
query8 = """
SELECT Number_of_Vehicles,
       AVG(CASE WHEN Accident_Severity = 'Serious' THEN 1 ELSE 0 END) AS serious_accident_rate
FROM accidents
GROUP BY Number_of_Vehicles
ORDER BY Number_of_Vehicles;
"""
vehicles_severity_correlation = pd.read_sql_query(query8, conn)
vehicles_severity_correlation
````

9. Occurrence of Accidents in Urban vs Rural Areas

````python
query9 = """
SELECT Urban_or_Rural_Area, COUNT(*) AS count
FROM accidents
GROUP BY Urban_or_Rural_Area
ORDER BY count DESC;
"""
urban_rural_occurrence = pd.read_sql_query(query9, conn)
urban_rural_occurrence
````
